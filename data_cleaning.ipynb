```python
import pandas as pd
import numpy as np
```

```python
country_mapping = pd.read_csv('./120-years-of-olympic-history-athletes-and-results/noc_regions.csv')
```

```python
country_mapping = country_mapping.drop(['notes'], axis = 1)
```

```python
total = pd.read_csv('./120-years-of-olympic-history-athletes-and-results/athlete_events.csv')
```

```python
athlete = total[['ID', 'Name', 'Sex', 'NOC']]

athlete = athlete.drop_duplicates()
```

```python
athlete['new_ID'] = np.arange(1,len(athlete)+1)
```

```python
ID_dict = athlete.set_index(['ID','Name','Sex','NOC']).to_dict()['new_ID']
total['new_ID'] = total.apply(lambda x: ID_dict[(x['ID'], x['Name'], x['Sex'], x['NOC'])] ,axis = 1)
```

```python
game = total[['Year', 'Season', 'City']]
```

```python
game = game.drop_duplicates()
```

```python
game['Game_ID'] = np.arange(len(game))
```

```python
gameID_dict = game.set_index(['Year','Season','City']).to_dict()['Game_ID']
```

```python
total['Game_ID'] = total.apply(lambda x: gameID_dict[(x['Year'], x['Season'], x['City'])] ,axis = 1)
```

```python
award = total[['new_ID', 'Game_ID', 'Event', 'Medal']]
award = award[award['Medal'].notnull()]
award = award.drop_duplicates()
```

```python
award = award[~award[['new_ID', 'Game_ID', 'Event']].duplicated()]
```

```python
NoAward = total[['new_ID', 'Game_ID', 'Event', 'Medal']]
NoAward = NoAward[NoAward['Medal'].isnull()]
```

```python
NoAward = NoAward[['new_ID','Game_ID','Event']]
NoAward = NoAward.drop_duplicates()
```

```python
sport_mapping = total[['Event','Sport']]
```

```python
sport_mapping = sport_mapping.drop_duplicates()
```

```python
country_mapping.to_csv("CountryMapping.csv", index = False)
athlete.to_csv("Athlete.csv", index = False)
game.to_csv("Game.csv", index = False)
award.to_csv("Award.csv", index = False)
NoAward.to_csv("NoAward.csv", index = False)
sport_mapping.to_csv("SportMapping.csv", index = False)
```

```python
athlete[['new_ID','ID','Name','Sex','NOC']].to_csv("Athlete.csv", index = False)
award[['new_ID','Game_ID','Event','Medal']].to_csv("Award.csv", index = False)
country_mapping[['NOC','region']].to_csv("CountryMapping.csv", index = False)
game[['Game_ID','Year','Season','City']].to_csv("Game.csv", index = False)
NoAward[['new_ID','Game_ID','Event']].to_csv("NoAward.csv", index = False)
sport_mapping[['Event','Sport']].to_csv("SportMapping.csv", index = False)
```
